#### tách dữ liệu 'Salary' ra 'max_salary' và 'min_salary'

```python
max=[]
min=[]
from_sal = re.compile("(Từ)")
to_sal = re.compile("(Tới)")
deal = re.compile("(Thương lượng)")
for i in df['salary']:
  if (from_sal.match(i)):
    max.append(np.nan)
    min.append(int(i.replace('Từ $','')))
  elif (to_sal.match(i)):
    min.append(np.nan)
    max.append(int(i.replace('Tới $','')))
  elif (deal.match(i)):
    max.append(np.nan)
    min.append(np.nan)
  elif (i=='nan'):
    max.append(np.nan)
    min.append(np.nan) 
  else:
    a,b=i.split(' - ')
    min.append(int(a.replace('$','')))
    max.append(int(b.replace('$','')))
df['min_salary']=min
df['max_salary']=max
```

```python
def current_change(money):
    if money > 1000000:
        money = money//23447
    return money
```

```python
def find_exp_year(df,exp):
    exp_year =[]
    for x in df:
        try:
            match= exp.findall(x)
            if len(match):
                lst=[]
                for x in match:
                    lst.append(x[0])
                exp_year.append(lst)
            else:
                exp_year.append([])
        except Exception as e:
            exp_year.append([])
    return exp_year
```

```python
exp_year1 = find_exp_year(df['requirement'],exp1)
exp_year1
```

```python
def find_exp_year_s2(exp_year,exp):
    least_lst=[]
    for x in exp_year:
        if x != []:
            row_lst=[]
            for i in x:
                match= exp.findall(i)
                lst=[]
                for a in match:
                    lst.append(float(a[0].replace(',','.')))
                row_lst.append(lst)
            row_lst = set(tuple(x) for x in row_lst)
            row_lst =[list(x) for x in row_lst]    
            least_lst.append(row_lst)
        else:
            least_lst.append([])
    return least_lst
```

```python
def mapping(col,map_df):
    map_lst=[]
    for x in col:
        elems=[]
        for i in x:
            elems.append(map_df.loc[map_df['name'] == i]['mapped_id'].tolist()[0])
        map_lst.append(sorted(list(set(elems))))
    return map_lst
```
